```python
import pandas as pd
from pandas.io.json import json_normalize
import json
from requests_html import HTMLSession
session = HTMLSession()
```

```python
metrika_headers = {
    'GET': '/management/v1/counters HTTP/1.1',
    'Host': 'api-metrika.yandex.net',
    'Authorization': 'It is a big secret', # здесь должен быть токен аутентификации, но свой я вам не скажу
    'Content-Type': 'application/x-yametrika+json'
    }

params = {
    'date1': '2022-01-01',
    'date2': '2022-05-31',
    'ids': 40599470
}
```

```python
url =  'https://api-metrika.yandex.net/stat/v1/data?ids=40599470'
# avgPageViews
report1 = '&metrics=ym:s:avgPageViews&dimensions=ym:s:operatingSystem'

report2 = 'https://api-metrika.yandex.net/stat/v1/data?preset=sources_summary&id=40599470'

#metrika_url = url + report1
metrika_url = report2
```

```python
json_content_response = session.get(metrika_url, headers=metrika_headers, params=params).json()
```
